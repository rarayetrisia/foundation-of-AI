```python
# Import library
import random
import nltk
import string
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
import warnings
import re
warnings.filterwarnings('ignore')

from gtts import gTTS
import IPython.display as ipd
import time
from io import BytesIO
```

Membuka dataset yang digunakan sebagai basic pengetahuan chatbot

```python
df = pd.read_csv('pertanyaanmahasiswa.txt', header=None, delimiter='\t')
df
```

Mengekstrak feature menggunakan TF-IDF

```python
#Extract Feature

from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer(lowercase=True)
X = vec.fit_transform(df[0])
question = X.toarray()
```

Pembuatan Chatbot

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
print('Bot : Selamat datang di Sistem Informasi Layanan Administrasi Universitas Malikusshaleh. Apa yang bisa kami bantu? Silahkan ajukan pertanyaan anda di bawah ini!') 
tts = gTTS('Selamat datang di Sistem Informasi Layanan Administrasi Universitas Malikusshaleh. Apa yang bisa kami bantu? Silahkan ajukan pertanyaan anda di bawah ini!', lang='id')
tts.save('pembukaan.wav')
            
exit_list = ['exit', 'sampai jumpa', 'close', 'tutup', 'bye', 'quit', 'break', 'stop']

while True:
  text = input('Mahasiswa : ').lower()
  if text in exit_list:
    print('Bot: Terimakasih telah mengunjungi layanan kami, semoga jawaban kami bisa menjawab pertanyaan anda :)')
    tts = gTTS('Terimakasih telah mengunjungi layanan kami, semoga jawaban kami bisa menjawab pertanyaan anda', lang='id')
    tts.save('terimakasih.wav')
    time.sleep(0.08)
    ipd.display(ipd.Audio('terimakasih.wav', autoplay=True))
            
    break
  else:
    text_vec = vec.transform([text])
    text_vec = text_vec.toarray()
    text_vec = text_vec[0].reshape(1, -1) 
    similarity = cosine_similarity(text_vec, question)
    answer_index = np.argmax(similarity[0])
    answer = df[1].iloc[answer_index]
    tts = gTTS(answer, lang='id')
    tts.save('jawaban.wav')
    time.sleep(0.08)
    print('Bot: ', answer)
    ipd.display(ipd.Audio('jawaban.wav', autoplay=True))
            
    
```
