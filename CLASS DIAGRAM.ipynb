```python
class departement:
  def __init__(self, Nama, angkatan, bidang):
    self.Nama = Nama
    self.angkatan = angkatan
    self.bidang = bidang
  def kontribusi(self):
    print(f'anggota angkatan {self.angkatan} {self.Nama}, berkontribusi aktif dalam bidang, {self.bidang}')    
  def proker(self):
    print(f"Proker bidang, {self.bidang}, telah selesai" )
  def sertifikat(self):
    print("cetak sertifikat", self.bidang)

anggota = departement('RARA YETRISIA PUTRI', 15, 'KPSDM')
anggota.kontribusi()
anggota.proker()
anggota.sertifikat()
```
