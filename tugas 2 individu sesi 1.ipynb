```python
# 3. prisma
la = 2
ka = 3
a = 4
t = 3
L = (2 * la)+(ka*t)
V = la * t
print("L = (2 * la)+(ka*t)\nV = la * t\n")
print(f"volume = {V}\nLuas= {L}")
```

```python
# 4. Limas
La = 4
t = 2
a_segitiga = 2
L = La + 4*(0.5 * a_segitiga * t)
V = 0.5 * La * t
print("L = La + 4(0.5 * a * t)\nV = 0.5 * La * t\n")
print(f"volume = {V}\nLuas= {L}")
```
